```python
import pandas as pd
#fetching raw data file
raw_dataset=pd.read_csv('/content/Frailty_raw_data.csv')
```

```python
#printing raw data
raw_dataset
```

```python
#preprocessing the data in preparation for analysis
 #updating the column names with the relevant units of measurement
raw_dataset.columns=['Height(in)', 'Weight(lb)', 'Age', 'Grip_strength(kg)','Frailty']
raw_dataset.columns

# Converting categorical data to numerical class values
raw_dataset['Frailty'].replace(['Y', 'N'],
                        [0, 1], inplace=True)
```

```python
clean_data=raw_dataset.dropna()
clean_data
```

```python
#Save the clean data to a csv file
clean_data.to_csv('/content/clean_data.csv')
print('successfully saved the data')
```
